```python
# ===============================
# üß† Import your RAG system
# ===============================
from my_rag import RAGSysstem  # Make sure this matches the uploaded filename

# ===============================
# üîê Enter your Gemini API Key
# ===============================
import getpass
API_KEY = getpass.getpass("AIzaSyDVxsY5RFOGkJYvgQ9e_QzG9rYePEYgLYw")

# ‚úÖ Initialize RAG system
rag = RAGSysstem(api_key=API_KEY)

# ===============================
# üìò Gradio chatbot functions
# ===============================
import gradio as gr

def upload_pdf_gradio(file):
    if file is None:
        return "No file uploaded."

    success = rag.upload_pdf(file.name)
    return "‚úÖ File uploaded and processed." if success else "‚ùå Failed to process file."

def ask_question_gradio(question):
    if not question.strip():
        return "Please enter a valid question."

    return rag.ask(question)

# ===============================
# üé® Gradio UI
# ===============================
with gr.Blocks() as demo:
    gr.Markdown("# üìò PDF Q&A Chatbot with Gemini")

    with gr.Row():
        file_input = gr.File(label="üìé Upload PDF", file_types=[".pdf"])
        upload_btn = gr.Button("üì§ Upload PDF")
    upload_output = gr.Textbox(label="Upload Status")
    upload_btn.click(upload_pdf_gradio, inputs=file_input, outputs=upload_output)

    gr.Markdown("---")

    question_box = gr.Textbox(label="‚ùì Ask a Question")
    answer_box = gr.Textbox(label="üí¨ Answer", lines=5)
    ask_btn = gr.Button("üîç Ask")

    ask_btn.click(ask_question_gradio, inputs=question_box, outputs=answer_box)

# ===============================
# üåê Launch Gradio app
# ===============================
demo.launch(share=True)
```
